```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
base_url='https://books.toscrape.com/catalogue/page-{}.html'
```

```python
page_number=1
while True :
    url=base_url.format(page_number)
    response=requests.get(url)
    if response.status_code!=200:
        break
    html_content=response.content
    
    soup=BeautifulSoup(html_content,'html.parser')
    library=soup.find_all('article',{'class':"product_pod"})
    for book in library:
        
        title=book.h3.a['title']
        price=book.find('p',{'class':'price_color'}).text.strip()
        rating='no rating' if 'star-rating' not in str(book) else book.find('p',{'class':'star-rating'})['class'][-1]
        print(f'Title:{title}')
        print(f'price:{price}')
        print(f'Rating:{rating}')
        print('=============================')
        
        book_file=open("library.txt","a",encoding='utf-8')
        book_file.write(f"{title} {price} {rating}\n")
        book_file.write("==================\n")
        
page_number +=1
        
```
